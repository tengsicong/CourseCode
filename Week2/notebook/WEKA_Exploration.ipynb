## Speculate About Design

We know that WEKA obtains its data from ARFF files, so there must be an ARFF file reader somewhere in the source directory structure (weka-3.8/weka/src/main). 

```bash
for file in `find weka-3.8/weka/src/main -name '*.java'`; 
do     
    echo "$file"; 
done | grep -i arff
```

We'll assume it's ./weka/core/converters/ArffLoader.java. If we have a look inside, we may find out what the ARFF contents are read into...

```bash
head -50 weka-3.8/weka/src/main/java/weka/core/converters/ArffLoader.java
```

```bash
for file in `find weka-3.8/weka/src/main -name '*.java'`; do
    echo "$file"; 
done | grep -i classifier
```

weka-3.8/weka/src/main/java/weka/classifiers/Classifier.java appears to be the abstract class / interface that represents all classifiers. Let's have a look inside...

```bash
head -100 weka-3.8/weka/src/main/java/weka/classifiers/Classifier.java
```

## Find Exceptional Entities

For each Java file, calculate the number of lines of code, and pipe the result to a CSV file in the tmp directory.
Once this is created, you can explore it with your favourite charting tool (e.g. Microsoft Excel).

```bash
for file in `find weka-3.8/weka/src/main -name '*.java'`do
    total=$(wc -l < $file)    
    echo "$file,$total"
done > /tmp/exceptionalEntities.csv

head -50 /tmp/exceptionalEntities.csv
```
